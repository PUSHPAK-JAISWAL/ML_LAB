Regression Analysis and Plot Interpretation

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
diabetes_X,diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
diabetes_X = diabetes_X[:, np.newaxis, 2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_X, diabetes_y, test_size=0.2, random_state=42
)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
plt.scatter(X_test, y_test, color="black", label="Actual Data")
plt.plot(X_test, y_pred, color="blue", linewidth=3, label="Regression Line")
plt.xlabel("BMI (Normalized)")
plt.ylabel("Disease Progression")
plt.title("Linear Regression: BMI vs Disease Progression")
plt.legend()
plt.show()
```

```python
print(f"Coefficient (Slope): {regr.coef_[0]:.2f}")
print(f"Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred):.2f}")
print(f"Coefficient of Determination (R²): {r2_score(y_test, y_pred):.2f}")
```
