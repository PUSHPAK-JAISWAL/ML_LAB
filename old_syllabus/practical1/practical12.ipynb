```r
data(mtcars)
```

```r
model <- lm(mpg ~ hp, data = mtcars)

summary(model)
```

```r
plot(mtcars$hp, mtcars$mpg, 
     pch = 16, 
     col = "steelblue", 
     cex = 1.2,
     main = "Regression Analysis: MPG vs Horsepower",
     xlab = "Horsepower (hp)", 
     ylab = "Miles Per Gallon (mpg)")

abline(model, col = "firebrick", lwd = 3)

grid(nx = NULL, ny = NULL, col = "lightgray", lty = "dotted")

par(mfrow = c(2, 2))
plot(model)
par(mfrow = c(1, 1))
```
