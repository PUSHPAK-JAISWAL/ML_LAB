Lab 7 Ad. Sale Prediction from Existing customer - Logistic Regression

```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
rows = 400

age = np.random.randint(18,60,rows)
salary = np.random.randint(15000, 150000, rows)

logit = 0.05 * age + 0.00005 * salary - 8
probability = 1 / (1 + np.exp(-logit))
purchased = (np.random.rand(rows) < probability).astype(int)

dataset = pd.DataFrame({
    'Age': age,
    'Salary': salary,
    'Status': purchased
})
```

```python
print(dataset.shape)
print(dataset.head())
```

```python
X = dataset.iloc[:, :-1].values
X
```

```python
Y = dataset.iloc[:, -1].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 0)
model.fit(X_train, y_train)
```

```python
age = int(input("Enter New Customer Age: "))
print(f"Enter New Customer Age: {age}")
sal = int(input("Enter New Customer Salary: "))
print(f"Enter New Customer Salary: {sal}")
newCust = [[age,sal]]
result = model.predict(sc.transform(newCust))
print(result)
if result == 1:
  print("Customer will Buy")
else:
  print("Customer won't Buy")
```

```python
y_pred = model.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
