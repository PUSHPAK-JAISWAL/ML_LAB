Implement and demonstrate the FIND-S algorithm for finding the most specific hypothesis based on a given set of training data samples. Read the training data from a .CSV file

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data_content = {
    'sky': ['sunny', 'rainy', 'rainy', 'sunny'],
    'airtemp': ['warm', 'warm', 'cold', 'warm'],
    'humidity': ['normal', 'high', 'high', 'high'],
    'wind': ['strong', 'strong', 'strong', 'strong'],
    'water': ['warm', 'warm', 'warm', 'cool'],
    'forcast': ['same', 'same', 'change', 'change'],
    'enjoysport': ['yes', 'yes', 'no', 'yes']
}
df_dataset = pd.DataFrame(data_content)
df_dataset.to_csv('enjoysport1.csv', index=False)

print("enjoysport1.csv created successfully:")
print(df_dataset)
```

```python
d = np.array(df_dataset)[:,:-1]
print("The attributes are: ",d)
```

```python
target = np.array(df_dataset)[:,-1]
print("n The target is: ",target)
```

```python
data = df_dataset
```

```python
target = np.array(data)[:,-1]
print("n The target is: ",target)
```

```python
def train(c,t):
    for i, val in enumerate(t):
        if val == "yes":  
            specific_hypothesis = c[i].copy()
            break

    for i, val in enumerate(c):
        if t[i] == "yes":
            for x in range(len(specific_hypothesis)):
                if val[x] != specific_hypothesis[x]:
                    specific_hypothesis[x] = '?'
                else:
                    pass

    return specific_hypothesis
```

```python
print("The final hypothesis is:",train(d,target))
```
